##Prompt template, Chat model, Output parser連結する例です。



```python
import os
from google.colab import userdata

os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")

from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "ユーザが入力した料理のレシピを考えてください。"),
        ("human", "{dish}")
    ]
)

model = ChatOpenAI(model_name="gpt-5.4-nano", temperature = 0)

chain = prompt | model | StrOutputParser()
output = chain.invoke({"dish": "カレー"})
print(output)
```

```python
#連結を使わない際の実装例

prompt_value = prompt.invoke({"dish": "カレー"})
ai_message = model.invoke(prompt_value)
output = StrOutputParser().invoke(ai_message)
print(output)
```

```python
#streamを使わう例
chain = prompt | model | StrOutputParser()

for chunk in chain.stream({"dish": "カレー"}):
    print(chunk, end="", flush=True)
```

```python
#batchメソッドを使うと、複数の入力をまとめて処理することもできる

chain = prompt | model | StrOutputParser()

outputs = chain.batch([{"dish":"カレー"},{"dish":"うどん"}])
print(outputs)
```

##chainとchainを連結する例

```python
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate

model = ChatOpenAI(model_name="gpt-5.4-nano", temperature = 0)
parser = StrOutputParser()

#一つ目のChain
thinking_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "ユーザの質問にステップ・バイ・ステップで考えて、回答してください。"),
        ("human", "{question}")
    ]
)
thinking_chain = thinking_prompt | model | parser

#2つ目のChain
summarize_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "ステップ・バイ・ステップで考えて回答から結論だけを抽出してください。"),
        ("human", "{question}")
    ]
)
summarize_chain = summarize_prompt | model | parser

final_chain = thinking_chain | summarize_chain
outputs=final_chain.invoke({"question": "10*2+3*2"})

print(outputs)


chain = prompt | model | StrOutputParser()
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful assistant."),
        ("human", "{input}")
    ]
)

model = ChatOpenAI(model_name="gpt-5.4-nano", temperature = 0)
parser = StrOutputParser()

#任意関数をRunnableにする
def upper(text:str) -> str:
    return text.upper()

#Runnableと任意の関数を「|」で接続すると、その関数を自動的にRunnableLambdaへ自動変換します
chain = prompt | model | parser | upper

output = chain.invoke({"input": "Hello!"})
print(output)
```

##複数のRunnableを並列につなげる

```python
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate

model = ChatOpenAI(model_name="gpt-5.4-nano", temperature = 0)
parser = StrOutputParser()

#楽観的な意見を生成するChain
optimistic_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "あなたは常に楽観的な意見を生成します。"),
        ("human", "{topic}")
    ]
)
optimistic_chain = optimistic_prompt | model | parser


#悲観的な意見を生成するChain
pessimistic_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "あなたは常に悲観的な意見を生成します。"),
        ("human", "{topic}")
    ]
)
pessimistic_chain = pessimistic_prompt | model | parser

#並列につなげる
import pprint
from langchain_core.runnables import RunnableParallel

parallel_chain = RunnableParallel(
    {
        "optimistic":optimistic_chain,
        "pessimistic":pessimistic_chain
    }
)

output = parallel_chain.invoke({"topic": "AI"})
pprint.pprint(output)
```

```python
from operator import itemgetter

synthesize_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "あなたは客観的なAIです。{topic}意見をまとめてください。"),
        ("human", "楽観的な意見：{optimistic}\n悲観的な意見：{pessimistic}")
    ]
)

syntheszize_chain = (
    {
        "optimistic":optimistic_chain,
        "pessimistic":pessimistic_chain,
        "topic":itemgetter("topic")
    }
    | synthesize_prompt
    | model
    | parser
)

output = syntheszize_chain.invoke({"topic": "AI"})
print(output)
```
